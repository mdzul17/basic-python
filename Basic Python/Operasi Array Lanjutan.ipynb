Operasi Array Bag. 2

remove() = Mengapus elemen array berdasarkan nilai/value pada array

pop() = Menghapus elemen array berdasarkan posisi nilai/value pada array

sort() = Mengurutkan array berdasarkan nilai terkecil (Ascending)

reverse() = Mengurutkan array berdasarkan nilai terbesar (Descending)

clear() = Menghapus semua elemen array

```python
nama = []
jml_nama = int(input("Masukkan jumlah data yang akan ditambahkan: "))
for i in range(jml_nama):
  nama_input = input('masukan nama mahasiswa ')
  nama.append(nama_input)
  print(nama)
```

```python
nama.remove("Agus")
print(nama)
```

```python
nama.sort()
print(nama)
```

```python
nama.reverse()
print(nama)
```

```python
nama.pop(0)
```

```python
nama.clear()
print(nama)
```

```python
nilai = []
input_data = int(input("Masukkan Data Input"))
for i in range(input_data):
  input_nilai = int(input("Masukkan Nilai ke-%d :" %i))
  nilai.append(input_nilai)
```

```python
print(nilai)
```

```python
for m in range(len(nilai)):
  if nilai[m]>=80:
    print("Indeks Nilai = A")
  elif nilai[m]>=60:
    print("Indeks Nilai = B")
  elif nilai[m]>=40:
    print("Indeks Nilai = C")
  else: print("Indeks Nilai = D")
```
