```python
#Membuat program input sederhana dengan memanfaatkan menu

class Menu:
  def __init__ (self, input_menu):
    self.input_menu = input_menu
  def pilihMenu(input_menu):
    if input_menu == 1:
        input_jumlah = int(input("Masukkan Jumlah Data: "))
        for i in range(input_jumlah):
          input_mahasiswa = input("Masukkan Nama Mahasiswa")
          dt_mahasiswa.append(input_mahasiswa)
    elif input_menu == 2:
        input_jumlah = int(input("Masukkan Jumlah yang ingin dihapus"))
        for i in range(input_jumlah):
          dt_mahasiswa.pop(0)
    elif input_menu == 3:
        print(dt_mahasiswa)
    else: print("Inputan tidak ditemukan")
```
