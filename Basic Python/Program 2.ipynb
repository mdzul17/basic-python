Membuat program perhitungan operasi bangun datar antara lain menghitung luas dan keliling. 

```python

#Membuat kelas Bangun Datar
class BangunDatar:
  def __init__ (self, nilai=0): #Nilai variabel Nilai tidak perlu diisi di program utama karena sudah mendapatkan nilai default yaitu 0
    self.nilai = nilai #Agar variabel nilai dapat dipakai di seluruh kelas turunan Bangun Datar

#Membuat kelas child (turunan) Jajar Genjang dari kelas bangun datar   
class JajarGenjang(BangunDatar):
  def luasJG(self, sisi):
    self.nilai = sisi*sisi
    return print("Nilai Luas Jajar Genjang adalah {}".format(self.nilai))

  def kelilingJG(self, sisi):
    self.nilai = sisi*4
    return print("Nilai Keliling Jajar Genjang adalah {}".format(self.nilai)) 

class PersegiPanjang(BangunDatar):
  def luasPP(self, panjang, lebar):
    self.nilai = panjang*lebar
    return print("Nilai luas persegi panjang adalah {}".format(self.nilai))

  def kelilingPP(self, panjang, lebar):
    self.nilai = (panjang*2) + (lebar*2)
    return print("Nilai keliling persegi panjang adalah {}".format(self.nilai))

class Segitiga(BangunDatar):
  def luasSegitiga(self, alas, tinggi):
    self.nilai = 1/2*alas*tinggi
    return print("Nilai luas segitiga adalah {}".format(self.nilai))
  
  def kelilingSegitiga(self, sisi1, sisi2, sisi3):
    self.nilai = sisi1+sisi2+sisi3
    return print("Nilai keliling segitiga adalah {}".format(self.nilai))
```

```python
#memulai Program Utama

#Membuat fungsi menu
def tampilMenu():
  print("-------Kumpulan Menu--------")
  print("1. Jajar Genjang\n2. Persegi Panjang\n3. Segitiga\n\n\n0. Keluar")
  pilihMenu = int(input("Pilih Menu: "))
  return pilihMenu

menu = tampilMenu()

while menu != 0:
  if menu == 1:
    jajarGenjang = JajarGenjang() #inisialisasi Kelas Jajar Genjang
    sisi = int(input("Masukkan nilai sisi: "))
    jajarGenjang.luasJG(sisi)
    jajarGenjang.kelilingJG(sisi)
    
  elif menu == 2:
    persegiPanjang = PersegiPanjang() #inisialisasi Kelas Persegi Panjang
    panjang = int(input("Masukkan nilai Panjang: "))
    lebar = int(input("Masukkan nilai lebar: "))
    persegiPanjang.luasPP(panjang, lebar)
    persegiPanjang.kelilingPP(panjang, lebar)
    
  elif menu == 3:
    segitiga = Segitiga() #inisialisasi kelas Segitiga
    nAlas = int(input("Masukkan nilai alas: "))
    nTinggi = int(input("Masukkan nilai tinggi: "))
    for i in range(3):
      nSisi[i] = int(input("Masukkan nilai sisi ke-",i,": "))
    segitiga.luasSegitiga(alas=nAlas, tinggi=nTinggi)
    segitiga.kelilingSegitiga(sisi1=nSisi[0], sisi2=nSisi[1], sisi3=nSisi[2])
    
  menu = tampilMenu()
```
