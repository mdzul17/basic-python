Seluruh parameter (argumen) pada bahasa Python bersifat “passed by reference”. Artinya saat mengubah sebuah variabel, maka data yang mereferensi padanya juga akan berubah, baik di dalam fungsi, maupun di luar fungsi pemanggil. Kecuali jika melakukan operasi assignment yang akan mengubah reference parameter.

```python
    def ubah(list_saya):
        list_saya.append([1, 2, 3, 5])
        print('Nilai di dalam fungsi: {}'.format(list_saya))
     
    # Panggil fungsi ubah
    list_saya = [10, 20, 30]
    ubah(list_saya)
    print('Nilai di luar fungsi: {}'.format(list_saya))
```

```python
    def ubah(list_saya):
        "Deklarasi Variabel list_saya berikut hanya dikenali (berlaku) di dalam fungsi ubah"
        list_saya = [1, 2, 3, 4] 
        print ('Nilai di dalam fungsi: {}'.format(list_saya))
     
    # Panggil fungsi ubah
    list_saya = [10, 20, 30]
    ubah(list_saya)
    print('Nilai di luar fungsi: {}'.format(list_saya))
```

1.   Keyword Argument, yakni argumen yang disertai identifier atau nama parameter yang secara eksplisit disebutkan. Hal ini termasuk jika kita mengirimkan nilai melalui dictionary yang diawali dua tanda * (**). 
2.   Positional Argument, yakni argumen selain keyword argument. Jika kita mengirimkan variabel bersifat iterable, maka harus diawali tanda *



```python
#Keyword Argument

daftar(tanggal=1, bulan='Januari', tahun=2020)
daftar(**{'tanggal': 1, 'bulan': 'Januari', 'tahun'=2020})

kali(angka2=5, angka1=4)

#Positional Argument

kali(5,4)
daftar(1,'Januari',2020)
daftar(*(1, 'Januari', 2020))
```

```python
    def printinfo(*args, **kwargs):
        for a in args:
            print('argumen posisi {}'.format(a))
        for key, value in kwargs.items():
            print('argumen kata kunci {}:{}'.format(key, value))
     
     
    # Panggil printinfo
    printinfo()
    printinfo(1, 2, 3)
    printinfo(i=7, j=8, k=9)
    printinfo(1, 2, j=8, k=9)
    printinfo(*(2, 3), **{'i':7, 'j':8})
```

Turunan (Inheritance)

```python
    class Kalkulator:
        """contoh kelas kalkulator sederhana. anggap kelas ini tidak boleh diubah!"""
     
        def __init__(self, nilai=0):
            self.nilai = nilai
     
        def tambah_angka(self, angka1, angka2):
            self.nilai = angka1 + angka2
            if self.nilai > 9:  # kalkulator sederhana hanya memroses sampai 9
                print('kalkulator sederhana melebihi batas angka: {}'.format(self.nilai))
            
    class KalkulatorKali(Kalkulator):
        """contoh mewarisi kelas kalkulator sederhana"""
     
        def kali_angka(self, angka1, angka2):
            self.nilai = angka1 * angka2
            return self.nilai
```

```python
    kk = KalkulatorKali()
    
    kk.tambah_angka(5, 6)  # memiliki fitur tambah_angka karena mewarisi dari Kalkulator
```

```python
#Menimpa (Override) Fungsi pada Parent Class
class KalkulatorKali(Kalkulator):
  """contoh mewarisi kelas kalkulator sederhana"""
     
  def kali_angka(self, angka1, angka2):
      self.nilai = angka1 * angka2
      return self.nilai
     
  def tambah_angka(self, angka1, angka2):
      self.nilai = angka1 + angka2
      return self.nilai
```

```python
    kk = KalkulatorKali()
     
    b = kk.kali_angka(5, 6)  # fitur tambah_angka yang dipanggil milik KalkulatorKali
    print(b)

    k = Kalkulator()

    c = k.tambah_angka(5,6)
    print(c)
```
